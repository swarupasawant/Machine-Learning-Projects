```python
#linear regression
from sklearn.model_selection import train_test_split
```

```python
#converting string date to date-time
import datetime as dt
india_cases['date']=pd.to_datetime(india_cases['date'])
india_cases.head(3)
```

```python
#converting date-time to ordinal
india_cases['date']= india_cases['date'].map(dt.datetime.toordinal)
india_cases.head(3)
```

```python
#getting dependent variable and independet variable
x= india_cases['date']
y= india_cases['total_cases']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
import numpy as np
model.fit(np.array(x_train).reshape(-1,1),np.array(y_train).reshape(-1,1))
```

```python
india_cases.tail(3)
```

```python
y_pred=model.predict(np.array(x_test).reshape(-1,1))
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score

a=r2_score(y_test,y_pred)
```

```python
print(a)
```

```python
from sklearn.metrics import max_error
max_error(y_test, y_pred)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(x_test,y_pred)
```
